####5. Drop the "point" feature from the dataframe - its a dictionary, which is unsupported in a flat database

```python
#YOUR CODE HERE
df = df.dropna(how = 'all').drop('point', 1).drop('media_url', 1)

```

####5. Turn the 311 data into a database. You can use the pandas read_sql command (example [here)]('http://bit.ly/1EoP81v') or the db package from yHat to do this. 

```python
df
```

```python
con = sqlite3.connect("../data/test_db.sqlite")
pd.io.sql.write_frame(df, "SF_311", con)
```

###### 6. Display the first 5 rows of the DB

```python
pd.read_sql("""Select * from SF_311 LIMIT 5""", con)
```

####7. Show 10 of the sidewalk cleanings in the Haight Ashbury neighborhood

```python
pd.read_sql("""Select * from SF_311 where category LIKE '%Sidewalk Cleaning%' and neighborhood LIKE '%Haight Ashbury%' LIMIT 10""", con)
```

####8. Which type of complaint was most common? 

```python
pd.read_sql("""Select category, count(request_type) AS Times_requested from SF_311 Group by request_type Order by Times_requested DESC LIMIT 1""", con)
```

####9. Are there any neighborhoods where sidewalk cleaning requests aren't the most common? 

```python
db2 = pd.read_sql("""Select neighborhood, category, count(request_type) as TotRequests from SF_311 where category 
                    LIKE '%Street and Sidewalk Cleaning%' group by neighborhood, category""", con)
pd.io.sql.write_frame(db2, "CountRequest", con)
```

```python
pd.read_sql("""Select SF_311.neighborhood, SF_311.category, count(SF_311.request_type) AS NumRequests from SF_311 
                where SF_311.category != 'Street and Sidewalk Cleaning' group by SF_311.neighborhood, SF_311.category having 
                numRequests > (select TotRequests from CountRequest where CountRequest.neighborhood = SF_311.neighborhood)""", con)
```
